### 4. Separate the string to find groups of string
1. Separate numbers from CoinsInnTier1 and store it in Coin
2. Remove the number part from CoinsInnTier1 and keep only the description (Description)
3. Clean NaN or empty values.

```python
# Separate into Amount in 'Coin' and 'Description'
df_c1['Coin'] = df_c1['CoinsInnTier1'].str.extract(r'([\d,]+)').replace({',': ''}, regex=True)
# Separate only text (cut out numbers)
df_c1['Description'] = df_c1['CoinsInnTier1'].str.replace(r'[\d,]+', '', regex=True).str.strip()
```

### 6. Handle this data
1. Replace No Charge, No Charge after deductible and $ in Description columns = 0 in Coinsurance columns
2. Drop NaN in Decriptions columns

```python
# Step1: Replace 'No Charge', 'No Charge after deductible', '0%' in Description to Coinsurance is 0
df_c1.loc[df_c1['Description'].isin(['No Charge', 'No Charge after deductible', '$']), 'Coin'] = '0'

# Step2: Drop 'NaN' in Coinsurance columns
# df_c1 = df_c1[df_c1['Description'] != 'NaN'].reset_index(drop=True)
df_cleaned = df_c1.dropna()
```

```python
# Used to convert data values in columns to numeric type.
df_cleaned['Coin'] = pd.to_numeric(df_cleaned['Coin'], errors='coerce')
```

### 7. Find out Mean of Coinsurance
1. Find out Mean of Coinsurance each PlanId and Add new columns Mean Coinsurance for check each PlanId have how mean of Coinsurance
2. Check manual
3. Calculate the Mean of each PlanId and take the resulting Mean file and add it as a new Column after the Description column by inserting the reference value from the original PlanId
4. Check Coinsurance of StandardComponentId that you want

```python
# Step1: Find out Copayment > 0 for test find out mean value of PlanId
plans_with_copayment = df_cleaned[df_cleaned['Coin'] > 0]['StandardComponentId'].unique()
plans_with_copayment
```

```python
# Step2: Check manual
# Filtered PlanId is 31274WV0420031 (Example)
filtered_df = df_cleaned[df_cleaned['StandardComponentId'] == '31274WV0420031']

# Count all value of Coin
coin_count = filtered_df['Coin'].notna().sum()

# Find valure that not duplicate value in Coinsurance and number of value
coin_value_counts = filtered_df['Coin'].value_counts().reset_index()
coin_value_counts.columns = ['Coin', 'Count']

# Calculate Coinsurance
coin_mean = filtered_df['Coin'].mean()

# Display recheck mean value of 31274WV0420031
print(f"all Coinsurance values: {coin_count}")
print("Values in Coin:")
print(coin_value_counts)
print(f"Mean of Coinsurance: {coin_mean:.2f}")

#Check manual 31274WV0420031 = ((20x42)+(50x2))/62 = 15.16 Mean
```

```python
# Step3: Calculate Mean of Copayment for each PlanId
mean_coin_per_plan = df_cleaned.groupby('StandardComponentId')['Coin'].mean().reset_index(name='Coinsurance')

# Merge the Mean value back into the original DataFrame
df_with_mean = pd.merge(df_cleaned, mean_coin_per_plan, on='StandardComponentId', how='left')

# Reorder columns to be after Description
columns_order = df_cleaned.columns.tolist() + ['Coinsurance']
df_with_mean = df_with_mean[columns_order]

# Display first 5 rows
df_with_mean.head()
```

```python
# display all columns, Non-Null Count and Dtype
df_with_mean.info()
```

```python
# Step4: Check Copayment of PlanId that you want
mean_value = df_with_mean.loc[df_with_mean['StandardComponentId'] == '31274WV0420031', 'Coinsurance'].iloc[0]

# Display
print(f"Coinsurance value of PlanId 31274WV0420031 is: {mean_value:.2f}")
```
